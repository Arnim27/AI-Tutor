```python
!pip install langchain faiss-cpu sentence-transformers transformers torch

from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.docstore.document import Document
import pickle

# Create embeddings
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# Sample documents
docs = [
    Document(page_content="This is the first document."),
    Document(page_content="This is the second document."),
    Document(page_content="Another document here.")
]

# Function to create vector store
def create_vector_store(docs):
    db = FAISS.from_documents(docs, embeddings)
    db.save_local("knowledge/faiss_index")
    return db

# Load vector store
def load_vector_store():
    return FAISS.load_local(
        "knowledge/faiss_index",
        embeddings,
        allow_dangerous_deserialization=True
    )

# Example usage
db = create_vector_store(docs)
print("✅ Vector store created and saved locally")
```
